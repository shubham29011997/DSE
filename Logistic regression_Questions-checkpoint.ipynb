```python
scale = StandardScaler()
```

```python
# df_target = pd.get_dummies(df1['Creditability'],drop_first=True)
targ = df1['Creditability']
df_ = df1.drop('Creditability',axis = 1)
# df_ = sm.add_constant(df_)
num = scale.fit_transform(df_)

# x_train,x_test,y_train,y_test = train_test_split(num,target,random_state = 10, test_size = 0.05)
# x_train.shape,y_train.shape

dfnum = pd.DataFrame(data = num,columns= df_.columns)
dfnum
resp = sm.add_constant(dfnum)
targ.info()
```

```python
x_train,x_test,y_train,y_test = train_test_split(resp,targ,random_state = 10, test_size = 0.05)
y_train
```

```python
log = sm.Logit(y_train,x_train).fit()
log.summary()
```

```python
ypred = log.predict(x_test)
y_pred = [ 0 if x < 0.5 else 1 for x in ypred]
metrics.accuracy_score(y_test,y_pred)
```

```python
cm = confusion_matrix(y_test,y_pred)
conf_matrix = pd.DataFrame(data = cm,columns = ['Predicted:0','Predicted:1'], index = ['Actual:0','Actual:1'])
conf_matrix
```

```python
acc = classification_report(y_test,y_pred)
print(acc)
```
